### Petite introduction

Le webscraping permet de collecter les données sur le web, pour effectuer une analyse des données.IL y'a plusieurs techniques.Nous utiliserons ici beautifulSoup.

j'ai saisi le lien de bbc news.

 # importer les packages

```python
# Code optionnel
from warnings import filterwarnings
filterwarnings("ignore")


import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# saisir l'URL

url="https://feeds.bbci.co.uk/news/rss.xml"

```

```python
# importer le code de la page
response = requests.get(url)

soup=BeautifulSoup(response.text, "html.parser")
soup
```

```python
# Inspection de ce qu'il à l'intérieur de item de toute une liste

items=soup.findAll("item")
```

```python
items
```

```python
# Inspection d'un seul élément dans items

items[0]
```

```python
# pour manipuler facilement, j'affecte ca à la variable item
item=items[0]
```

```python
# Extraire uniquement le titre dans cet élément

item.title.text
```

```python
# si on veut niquement la description

item.description.text
```

```python
# création d'une liste vide et affectation à la variable news_ites

news_items= []
```

```python
# boucle for
for i in items: #pour tout i dans la liste items
    news_i={}   # création du dictionnaire {} puis on affecte à la variable news_i
    news_i["title"]=i.title.text
    news_i["description"]=i.description.text
    news_i["pubdate"]=i.pubdate.text
    
    
    news_items.append(news_i)
    
    
    
```

```python
# recupération d'élément après élément("title", "description","pubdate" : c'est la date)

news_items # Le sortie des informationssontsous la forme de liste de dictionnaire [{}]
```

```python
# Convertir la liste en base de données

df=pd.DataFrame(news_items, columns=["title", "description", "pubdate"])
```
